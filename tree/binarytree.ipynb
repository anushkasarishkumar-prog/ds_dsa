```python
def inorder(root):                                           #inorder
    if root==None:
        return
    inorder(root.left)
    print(root.data)
    inorder(root.right)
inorder(root)
```

```python
def preorder(root):
    if root==None:
        return
    print(root.data)
    preorder(root.left)
    preorder(root.right)
preorder(root)
```

```python
def postorder(root):
    if root==None:
        return
    preorder(root.left)
    preorder(root.right)
    print(root.data)
    
postorder(root)
```

```python
class Tree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def insert(root, value):
    if value < root.data:
        if root.left is None:
            root.left = Tree(value)
        else:
            insert(root.left, value)
    else:
        if root.right is None:
            root.right = Tree(value)
        else:
            insert(root.right, value)

```
