```python
class Customer:
    def __init__(self,id,name,address):
        self.id=id
        self.name=name
        self.address=address
    def std_info(self):
        print(f"id:{self.id} name:{self.name} address{self.address}")
```

```python
c=Customer(100,"joe","bengalore")
```

```python
class Cart:
    def __init__(self,name,code,price,qty):
        self.name=name
        self.code=code
        self.price=price
        self.qty=qty
    def cart_info(self):
        total=self.price*self.qty
        print(f"name:{self.name} code:{self.code} price:{self.price} qty:{self.qty} total:{total}")
```

```python
cart=Cart("bun",123,12,31)
```

```python
cart.cart_info()
```
