```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
    
```

```python
n1=Node(10)
n2=Node(20)
n3=Node(30)
n4=Node(40)
print(n1.data)
print(n2.data)
print(n3.data)
print(n4.data)
```

```python
n1.next=n2
n2.next=n3
n3.next=n4
n4.next=None
```

```python
cur=n1
while cur:
    print(cur.data)
    cur=cur.next
```

```python
#a=print(n1.data+n2.data+n3.data+n4.data)
s=0
cur=n1
while cur:
    s+=cur.data
    cur=cur.next
print(s)
```

```python
#print(max(n1.data,n2.data,n3.data,n4.data))
cur=n1
max=0
while cur:
    if cur.data>max:
        max=cur.data
    cur=cur.next
print(max)
```

```python
class Student:
    def __init__(self,id,name,result):
        self.id=id
        self.name=name
        self.result=result
    def Student_info(self):
        
        print(f"Student_id:{self.id}    Student_name:{self.name}    Student_result:{self.result}")
```

```python
n=int(input("enter how many students"))
students=[]
```

```python
for i in range(n):
    id=int(input("enter student id"))
    name=input("enter student name")
    result=input("result :")
    s1=Student(id,name,result)
    students.append(s1)
```

```python
for s1 in students:
    s1.Student_info()
```
