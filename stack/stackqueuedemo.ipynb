```python
class Stack:
    def __init__(self):
        self.data=[]
    def push(self,item):
        self.data.append(item)
    def display(self):
        #for i in range(-1,-6,-1):
           # print(self.data[i],sep=" ")
        #print(self.data.reversed)
        for i in range(-1,-len(self.data)+1,-1):
            print(self.data[i],sep=" ")
    def pop(self):
        if self.is_empty()!=True:
            return self.data.pop()
        else:
            print("Stack is empty")
    def is_empty(self):
        if len(self.data)==0:
            return True
    def peak(self):
        if self.is_empty()!=True:
            return self.data[-1]
        else:
            print("Stack is empty")
        
        #3 operations in stack push pop peak
        
```

```python
s=Stack()
s.push(10)
s.push(20)
s.push(30)
s.display()
s.pop()
s.pop()
s.pop()
s.pop()
s.peak()
```

```python
class Order:
    def __init__(self,order_id ,date,price,quantity):
        self.id=id
        self.date=date
        self.price=price
        self.quantity=quantity
    def order_info(self):
        print(f"Orderid:{self.id}OrderDate:{self.date}OrderPrice:{self.price}OrderQuantity:{self.quantity}")
    
```

```python
o1=Order(100,"10-10-2025",200,2)
o2=Order(102,"11-11-2025",900,10)
o3=Order(105,"10-1-2025",600,3)
o4=Order(107,"11-09-2025",700,5)
o5=Order(106,"22-01-2025",2000,7)
```

```python
 from collections import deque
```

```python
order_queue=deque()
```

```python
order_queue.append(o1)
order_queue.append(o2)
order_queue.append(o3)
order_queue.append(o4)
order_queue.append(o5)
```

```python
while order_queue:
    cur=order_queue.popleft()
    print("processing order")
    cur.order_info()
```

```python
class Patient:
    def __init__(self,id,name,dep,type):
        self.id=id
        self.name=name
        self.dep=dep
        self.type=type
    def patient_info(self):
        print("Patientid={self.id} Patientname={self.name} Department={self.dep}Type={self.type}")
```

```python
p1=Patient(101,"john","opd","inp")
p2=Patient(102,"joe","opd","outp")
p3=Patient(100,"ganga","ipd","inp")
p4=Patient(106,"marin","opd","outp")
p5=Patient(105,"elen","ipd","inp")
```

```python
import heapq
patients=[]
```

```python
heapq.heappush(patients,(3,p1))
heapq.heappush(patients,(2,p2))
heapq.heappush(patients,(1,p3))
heapq.heappush(patients,(4,p4))
heapq.heappush(patients,(5,p5))
```

```python
while patients:
    pr,p=heapq.heappop(patients)
    p.patient_info()
```
